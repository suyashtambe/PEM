Loads excel file into a pandas DataFrame.    

An excel file with data points of each state ( normal and faulty ) is taken.

```python
faulty=pd.read_excel("/content/drive/MyDrive/SEM 3/DPL/project/sensor measurement/faulty state 1.xls")
normal=pd.read_excel("/content/drive/MyDrive/SEM 3/DPL/project/sensor measurement/normal state 1.xls")
```

A new column called 'flag' to the 'faulty' DataFrame and assigns a value of 1 to all rows in this column. This 'flag' column is used to label the 'faulty' data.    
    
Similarly, a 'flag' column is added to the 'normal' DataFrame and assigned a value of 0 to all rows in this column. This 'flag' column is used to label the 'normal' data.


```python
faulty['flag']=1
normal['flag']=0
```

The 'faulty' and 'normal' DataFrames are combined into a new DataFrame called 'merge.' The result is a single DataFrame that contains data from both 'faulty' and 'normal' states. The 'flag' column helps differentiate between the two states, with 1 representing 'faulty' and 0 representing 'normal' data.

```python
merge=pd.concat([faulty,normal])
```

This line takes the 'merge' DataFrame and shuffles its rows.

frac=1 means that the the entire DataFrame is selected, shuffling all the rows in the DataFrame.
random_state=42 sets the random seed to 42, ensuring that the shuffling process is reproducible.

```python
shuffled_df = merge.sample(frac=1, random_state=42)
```

The StandardScaler class is imported from the scikit-learn library. The StandardScaler is used for standardizing features by removing the mean and scaling to unit variance.    

A copy of the original dataset is created for the application of standard scaling technique. The 'flag' is not included in the copy of the dataset to maintain the label values.      

A scaling object is created and then applied to the dataframe using
 .fit_transform().

The resulting array is then converted a dataframe.

```python
#importing the z score technique
from sklearn.preprocessing import StandardScaler
#creating a copy of the dataframe
scaled_df=df.iloc[:, :-1]
#creating an normalization object
zscore_scaler=StandardScaler()
After_scale = zscore_scaler.fit_transform(scaled_df)
After_scale_df=pd.DataFrame(After_scale,columns=scaled_df.columns)

After_scale_df
```

The LabelEncoder class is imported from the scikit-learn library.LabelEncoder is used for converting categorical labels into numerical labels.

The ranges and thier corresponding labels are defined for categorizing the numeric data.These ranges represent different categories, and the labels provide a human-readable description for each category.

```python
from sklearn.preprocessing import LabelEncoder

# Define the ranges and labels
bins = [0, 25, 50,  float('inf')]
labels = ['0-25', '25-50', '50+']
```

A new column is created in the DataFrame called 'category' by using the pd.cut() function. It takes the values from the 'current' column and categorizes them based on the specified 'bins' and assigns the corresponding 'labels' to each category. The right=False argument ensures that the intervals are left-inclusive and right-exclusive.

```python
# Use pd.cut to categorize the numeric column
df['category'] = pd.cut(df[' current'], bins=bins, labels=labels, right=False)
```

A label encoder object is intialized and then applied to the dataframe using
.fit_transform().

```python
# Initialize the LabelEncoder
label_encoder = LabelEncoder()
# Fit and transform the 'category' column to numeric labels
df['category_encoded'] = label_encoder.fit_transform(df['category'])
```
